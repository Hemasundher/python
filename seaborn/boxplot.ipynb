#box plot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

box plot 

```python
s.boxplot(y='Total',data=df, width=0.2)
```

boxplot and one categorical value

```python
s.boxplot(x='Payment',y='Total',data=df)
```

boxplot with one numerical and two categorical

```python
s.boxplot(x='Payment',y="Total",hue='Gender',data=df,hue_order=['Male','Female'])
```

adding mean to plot

```python
s.boxplot(x='Payment',y='Total',hue='Gender',data=df,showmeans=True,meanprops={"marker":"o",
                                                                                "markerfacecolor":"white",
                                                                                "markersize":"10",
                                                                                "markeredgecolor":"black",
                                                                                })
```

```python
s.boxplot(y='Payment',x='Total',hue='Gender',data=df,showmeans=True,meanprops={"marker":"o",
                                                                                "markerfacecolor":"white",
                                                                                "markersize":"10",
                                                                                "markeredgecolor":"black",
                                                                                })
```

palette and linewidth

```python
s.boxplot(y='Payment',x='Total',hue='Gender',data=df,showmeans=True,meanprops={"marker":"o",
                                                                                "markerfacecolor":"white",
                                                                                "markersize":"10",
                                                                                "markeredgecolor":"black",
                                                                                },
palette="CMRmap",
linewidth=1.4
)
```

to plot all numerics in the df

```python
plt.figure(figsize=(15,5))
s.boxplot(data=df)
```
