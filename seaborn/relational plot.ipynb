```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('mart_train_sample.csv').sample(50)
df.head()
```

```python
df.columns=df.columns.str.lower()
df.columns
```

```python
s.relplot(data=df,x='outlet_year',y='sales')
```

```python
s.relplot(data=df,x='outlet_year',y='sales',kind='line',errorbar=None)
```

```python
s.relplot(data=df,x='outlet_year',y='sales',s=100,hue='outlet_size',
            col='tier'
            )
```

```python
s.relplot(data=df,x='outlet_year',y='sales',s=100,hue='outlet_size',
            col='item_type'
            )
```

```python
s.relplot(data=df,x='outlet_year',y='sales',s=100,hue='outlet_size',
            col='item_type',
            col_wrap=5
            )
```

```python
s.relplot(data=df,x='outlet_year',y='sales',s=100,hue='outlet_size',
            col='outlet_size',
            row='tier'
            )
```

```python
s.relplot(data=df,x='outlet_year',y='sales',s=100,hue='outlet_size',
            palette=['green','yellow','red']
            )
```
