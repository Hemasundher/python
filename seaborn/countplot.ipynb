```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

countplot() to product line

```python
plt.figure(figsize=(15,5))
s.countplot(x='Product line',data=df)
```

horizontal bar plot

```python
s.countplot(y='Product line',data=df)
```

adding hue to get count on two categories  -- product line and gender

```python
plt.figure(figsize=(15,5))
s.countplot(x='Product line',hue='Gender',data=df)
```

adding color pallete

```python
plt.figure(figsize=(15,5))
a=s.countplot(x='Product line',hue='Gender',data=df,palette="rainbow")

```

adding number to bars

```python
plt.figure(figsize=(15,5))
a=s.countplot(x='Product line',hue='Gender',data=df,palette="rainbow")


for i in a.containers:
    a.bar_label(i,)
```

face color,linewidth,edgecolor

```python
plt.figure(figsize=(15,5))
a=s.countplot(x='Product line',hue='Gender',data=df,
            facecolor=(1,0,0,0),
            linewidth=5,
            edgecolor=s.color_palette('dark',3),
            
            
            )


for i in a.containers:
    a.bar_label(i,)
```
