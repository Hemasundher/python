# joint plot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
iris=s.load_dataset('iris')
iris.head()
```

basic jointplot

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length')
```

kind - scatter,hist,hex,reg,resid,kde

```python
plt.figure(figsize=(15,5))

# plt.subplot(2,3,1)
s.jointplot(data=iris,x='sepal_length',y='petal_length',kind='scatter')
```

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',kind='hist')
```

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',kind='hex')
```

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',kind='reg')
```

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',kind='resid')
```

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',kind='kde')
```

using categorical value

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

formatting all and individual plots

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',
            # color='red',
            # hue='species',
            # palette='BuPu',

            joint_kws=dict(marker='+',color='red'),
            marginal_kws=dict(color='black',kde=True),


            )
```

adjust height , ratio ,space and show/hide marginal_ticks

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',height=5)
```

```python
s.jointplot(data=iris,x='sepal_length',y='petal_length',
                ratio=2,
                # space=3,
                marginal_ticks=True,
                )
```

kde and rug

```python
p=s.jointplot(data=iris,x='sepal_length',y='petal_length',joint_kws=dict(color='red'))
p.plot_joint(s.kdeplot)
p.plot_joint(s.rugplot,height=0.1,color='red')
```
