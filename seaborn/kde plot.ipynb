# kde plot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

```python
df=df[['Gender','Payment','Unit price','Quantity','Total','gross income']]
df.head()
```

basic kde plot

```python
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
s.kdeplot(data=df,x='Total')


plt.subplot(1,2,2)
s.kdeplot(data=df,x='gross income')
```

kde for all numeric variables

```python
s.kdeplot(data=df)
```

smoothing using bw_adjust

```python
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
s.kdeplot(data=df,x='Total',bw_adjust=0.1)

plt.subplot(1,2,2)
s.kdeplot(data=df,x='Total',bw_adjust=0.6)
```

kde - categorical value

```python
plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
s.kdeplot(data=df,x='Total',hue='Gender')

plt.subplot(2,2,2)
s.kdeplot(data=df,x='gross income',hue='Gender')

plt.subplot(2,2,3)
s.kdeplot(data=df,x='Total',hue='Payment')

plt.subplot(2,2,4)
s.kdeplot(data=df,x='gross income',hue='Payment')


```

stack of category kde

```python
s.kdeplot(data=df,x='Total',hue='Payment',multiple='stack')
```

log scaling

```python
s.kdeplot(data=df,x='Total',log_scale=True)
```

linewidth,palette

```python
s.kdeplot(data=df,x='Total',hue='Payment',multiple='stack',linewidth=1,palette='Dark2',alpha=0.2)
```

bivariate kde

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
s.kdeplot(data=df,x='Unit price',y='gross income')


plt.subplot(1,2,2)
s.kdeplot(data=df,x='gross income',y='Total')
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
s.kdeplot(data=df,x='Unit price',y='gross income',fill=True)


plt.subplot(1,2,2)
s.kdeplot(data=df,x='gross income',y='Total',fill=True)
```

bivariate on categorical value

```python
plt.figure(figsize=(15,5))

s.kdeplot(data=df,x='Unit price',y='gross income',hue='Gender',fill=True,levels=2)
```

```python
plt.figure(figsize=(15,5))

s.kdeplot(data=df,x='Unit price',y='gross income',hue='Gender',fill=True,levels=5,thresh=0.2)
```
