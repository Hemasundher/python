# histplot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

```python
df=df[['Gender','Payment','Unit price','Quantity','Total','gross income']]
df.head()
```

basic hist

```python
bins=n.linspace(0,1000,6)
s.histplot(data=df,x='Total',bins=bins)
```

change bins width using binwidth 

```python
s.histplot(data=df,x='Total',binwidth=160)
```

change no of bins and intervals

```python
s.histplot(data=df,x='Total',bins=20)
```

```python
s.histplot(data=df,x='Total',bins=n.arange(0,1100,50))
```

```python
plt.figure(figsize=(15,5))
s.histplot(data=df,x='Total',bins=n.arange(0,1100,50))
plt.xticks(n.arange(0,1100,50))
```

combine with kde

```python
s.histplot(data=df,x='Total',kde=True)
```

addign hue -- categorical value

```python
s.histplot(data=df,x='Total',hue='Payment')
```

```python
s.histplot(data=df,x='Total',hue='Payment',multiple='stack')
```

```python
plt.figure(figsize=(15,5))
s.histplot(data=df,x='Total',bins=n.arange(0,1100,50),hue='Payment',multiple='stack')
plt.xticks(n.arange(0,1100,50))
```

using step,poly

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
s.histplot(data=df,x='Total',element='step')
plt.subplot(1,2,2)
s.histplot(data=df,x='Total',element='poly')
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
s.histplot(data=df,x='Total',hue='Payment',element='step')
plt.subplot(1,2,2)
s.histplot(data=df,x='Total',hue='Payment',element='poly')
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
s.histplot(data=df,x='Total',hue='Payment',element='step',fill=False)
plt.subplot(1,2,2)
s.histplot(data=df,x='Total',hue='Payment',element='poly',fill=False)
```

histplot to categorical value and shrink 

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
s.histplot(data=df,x='Payment',shrink=0.7)
plt.subplot(1,2,2)
s.histplot(data=df,x='Gender',shrink=0.7)
```

stats - count ,density,probability,frequency

```python
plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
s.histplot(data=df,x='Total',stat='count')

plt.subplot(2,2,2)
s.histplot(data=df,x='Total',stat='density')

plt.subplot(2,2,3)
s.histplot(data=df,x='Total',stat='probability')

plt.subplot(2,2,4)
s.histplot(data=df,x='Total',stat='frequency')


```

bivariate histogram 

```python
s.histplot(data=df,x='Total',y='gross income')
```
