# lineplot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('mart_train_sample.csv').sample(50)
df.head()
```

basic line plot

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None)
```

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,estimator='sum')
```

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,estimator=None)
```

test with different confidence intervals and with different no of bootstraping

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',)
```

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar='sd')
```

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',n_boot=2)
```

hue 

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,hue='Outlet_Size',
                palette='rocket_r',
                )
```

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,hue='Outlet_Size',
                palette=['red','green','blue'],
                )
```

grouping using styles,use different markers and dashes

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,
            style='Outlet_Size',
            markers=True,
            dashes=False
                )
```

grouping usings size

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,
            size='Outlet_Size',
            sizes=(.5,2)
                )
```

semantic styling parameters on same different variables

```python
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,
            hue='Outlet_Size',
            style='Outlet_Size',
            markers=True
                )
```

```python
plt.figure(figsize=(10,5))
s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,
            hue='Outlet_Size',
            style='Tier',
            markers=True
                )
```

```python

s.lineplot(data=df,x='Outlet_Year',y='Sales',errorbar=None,
            hue='Outlet_Size',
            # style='Tier',
            # markers=True
            size='Tier',
                )
```
