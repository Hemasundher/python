# rugplot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

```python
df=df[['Gender','Payment','Unit price','Quantity','Total','gross income']]
df.head()
```

basic rogplot

```python
plt.figure(figsize=(15,5))
s.rugplot(data=df,x='Unit price')
```

two variables

```python
plt.figure(figsize=(15,5))
s.rugplot(data=df,x='Unit price',y='gross income')
```

group by hue

```python
plt.figure(figsize=(15,5))
s.rugplot(data=df,x='Unit price',y='gross income',hue='Gender')
```

height

```python
plt.figure(figsize=(15,5))
s.rugplot(data=df,x='Unit price',y='gross income',hue='Gender',height=0.09)
```

kdeplot

```python
plt.figure(figsize=(15,5))
s.rugplot(data=df,x='Unit price',y='gross income',hue='Gender')
s.kdeplot(data=df,x='Unit price',y='gross income',hue='Gender',fill=True)
```

scatter plot

```python
plt.figure(figsize=(15,5))
s.rugplot(data=df,x='Unit price',y='gross income',hue='Gender')
s.scatterplot(data=df,x='Unit price',y='gross income',hue='Gender')
```

outside

```python
plt.figure(figsize=(15,5))
s.rugplot(data=df,x='Unit price',y='gross income',hue='Gender',height=-0.01,clip_on=False)
s.scatterplot(data=df,x='Unit price',y='gross income',hue='Gender')
```
