# heatmap

```python
import pandas as p
import numpy as n
import matplotlib.pyplot as plt
import seaborn as s
```

```python

df=p.read_csv('mart_linePlot.csv')
df.columns=df.columns.str.lower()
df.head()
```

```python
df.shape
```

```python
df1=df.pivot_table(index='outlet_year',columns='outlet_size',values='sales')
df1
```

```python
s.set_style('white')
plt.figure(figsize=(5,5))
s.heatmap(df1)
```

```python
s.set_style('white')
plt.figure(figsize=(5,5))
s.heatmap(df1,annot=True)
```

```python
s.set_style('white')
plt.figure(figsize=(5,5))
s.heatmap(df1,annot=True,fmt='.0f',
            annot_kws=dict(size=15,weight='bold'),
            linewidths=0.5,
            linecolor='black',
            cmap='rocket_r'
            )
```

correlation


```python
df.corr()
```

```python
s.heatmap(df.corr())
```

```python
s.heatmap(df.corr(),
    vmin=-1,
    vmax=1,
    center=0,
    cmap='OrRd_r',
    annot=True,
    fmt='.1f',
    annot_kws=dict(size=15,weight='bold'),
    linecolor='black',
    linewidths=3
)
```
