# facetgrid plot

```python
import pandas as p
import numpy as n
import matplotlib.pyplot as plt
import seaborn as s
```

```python

df=p.read_csv('mart_linePlot.csv')
df.columns=df.columns.str.lower()
df.head()
```

```python
df['outlet_size'].unique()
```

```python
df['outlet_location_type'].unique()
```

```python
fg1=s.FacetGrid(df,col='outlet_size')
```

hist plot

```python
fg1=s.FacetGrid(df,col='outlet_size')
fg1.map_dataframe(s.histplot,x='sales')
```

```python
fg1=s.FacetGrid(df,col='outlet_size')
fg1.map_dataframe(s.kdeplot,x='sales')
```

```python
fg1=s.FacetGrid(df,col='outlet_size')
fg1.map_dataframe(s.lineplot,x='outlet_year',y='sales')
```

```python
fg1=s.FacetGrid(df,col='outlet_size')
fg1.map_dataframe(s.histplot,x='outlet_year',y='sales')
```

```python
fg1=s.FacetGrid(df,col='outlet_size')
fg1.map_dataframe(s.scatterplot,x='outlet_year',y='sales',
                    hue='outlet_location_type',
                    marker='+',
                    alpha=0.5,
                    color='red',
                    s=100
                    
                    
                    )
```

```python
fg2=s.FacetGrid(df,col='outlet_size',row='outlet_location_type')
```

```python
fg2=s.FacetGrid(df,col='outlet_size',row='outlet_location_type',sharey=False,ylim=(0,1000))
fg2.map_dataframe(s.histplot,x='sales')
fg2.set_axis_labels('sales','count')
fg2.set_titles(col_template='{col_name} size',row_template='{row_name} type')
```
