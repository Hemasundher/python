scatter plot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('mart_train_sample.csv').sample(50)
df.head()
```

basic scatter plot  - Sales vs Item_MRP	

```python
s.scatterplot(data=df,x='Item_MRP',y='Sales')
```

hue

```python

s.scatterplot(data=df,x='Item_MRP',y='Sales',hue='Outlet_Size')
```

styles for a categorical value

```python
s.scatterplot(data=df,x='Item_MRP',y='Sales',style='Outlet_Size',s=100)
```

```python
s.scatterplot(data=df,x='Item_MRP',y='Sales',style='Outlet_Size',
            markers={"High":"^","Small":"v","Medium":"o"}
            )
```

hue - shape and color 

```python

s.scatterplot(data=df,x='Item_MRP',y='Sales',style='Outlet_Size',
            markers={"High":"^","Small":"v","Medium":"o"},
            hue='Outlet_Size',
            )
```

```python

s.scatterplot(data=df,x='Item_MRP',y='Sales',style='Outlet_Size',
            markers={"High":"^","Small":"v","Medium":"o"},
            hue='Outlet_Location_Type',
            )
```

hue - palette

```python

s.scatterplot(data=df,x='Item_MRP',y='Sales',
            #style='Outlet_Size',
            markers={"High":"^","Small":"v","Medium":"o"},
            hue='Outlet_Size',
            palette='Purples_r'
            )
```

hue and size

```python

s.scatterplot(data=df,x='Item_MRP',y='Sales',
            size='Tier',
            markers={"High":"^","Small":"v","Medium":"o"},
            hue='Outlet_Size',
            )
```

```python

s.scatterplot(data=df,x='Item_MRP',y='Sales',
            size='Tier',
            sizes=(20,200),
            markers={"High":"^","Small":"v","Medium":"o"},
            hue='Outlet_Size',
            )
```

```python

s.scatterplot(data=df,x='Item_MRP',y='Sales',
            size='Tier',
            sizes=(20,200),
            legend='full',
            markers={"High":"^","Small":"v","Medium":"o"},
            hue='Outlet_Size',
            )
```

markersize - s

```python
s.scatterplot(data=df,x='Item_MRP',y="Sales",s=100,color='red',edgecolor='black')
```
