# ecdf - empirical cumulative distributive function

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

```python
df=df[['Gender','Payment','Unit price','Quantity','Total','gross income']]
df.head()
```

basic ecdf

```python
s.ecdfplot(data=df,x='gross income')
```

adding categoric value

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
s.ecdfplot(data=df,x='gross income',hue='Gender')

plt.subplot(1,2,2)
s.ecdfplot(data=df,x='gross income',hue='Gender',stat='count')
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
s.ecdfplot(data=df,y='gross income',hue='Gender')

plt.subplot(1,2,2)
s.ecdfplot(data=df,y='gross income',hue='Gender',stat='count')
```
