```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
tips=s.load_dataset('tips')
tips.head()
```

basic reg plot

```python
s.regplot(data=tips,x='total_bill',y='tip')
```

style of reg

```python
s.regplot(data=tips,x='total_bill',y='tip',color='red',marker='+')
```

```python
s.regplot(data=tips,x='total_bill',y='tip',line_kws={'color':'red','linestyle':'--'},
            
            scatter_kws={'color':'green','alpha':0.1},
            ci=60,
            n_boot=500
            )
```

```python
s.regplot(data=tips,x='total_bill',y='tip',logistic=True)
```
