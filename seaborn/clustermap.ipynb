# cluster map

```python
import pandas as p
import numpy as n
import matplotlib.pyplot as plt
import seaborn as s
```

```python

df=p.read_csv('mart_linePlot.csv')
df.columns=df.columns.str.lower()
df.head()
```

pivot creation

```python
df1=df.pivot_table(index='outlet_year',columns='outlet_size',values='sales')
df1
```

```python
s.clustermap(df1)
```

```python
s.clustermap(df1,col_cluster=False)
```

```python
s.clustermap(df1,row_cluster=False)
```

```python
s.clustermap(df1,annot=True,fmt='.0f',z_score=1)
```

```python
s.clustermap(df1,annot=True,fmt='.0f',standard_scale=0,linewidth=0.5)
```
