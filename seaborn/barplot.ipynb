```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

all datasets of seaborns

```python
s.get_dataset_names()
```

```python
data=s.load_dataset('diamonds')
data.head()
```

getting data 

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

```python
df['Product line'].info()
```

bar plot

```python
plt.figure(figsize=(15,5))
s.barplot(x='Product line',y='Total',data=df)
```

adding hue to plot

```python
plt.figure(figsize=(15,5))
s.barplot(x='Product line',y='Total',hue='Gender',data=df)
```

horizontal

```python
plt.figure(figsize=(15,5))
s.barplot(y='Product line',x='Total',hue='Gender',data=df)
```

order of bars in the plot

```python
l1=list(set(df['Product line'].to_list()))
l1.sort()
l1
```

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'])
```

add cap on error bar

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],capsize=0.2)
```

remove the error bar

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],errorbar=None)
```

add color

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],color='black')
```

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',data=df,order=l1,hue_order=['Male','Female'],color='black')
```

change color using palette


```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],palette='cividis')
```

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',data=df,order=l1,hue_order=['Male','Female'],palette='cividis')
```

use saturation

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],palette='cividis',saturation=0.5)
```

use estimator

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],palette='cividis',estimator=sum)
```

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],palette='cividis',estimator=n.median)
```

```python
plt.figure(figsize=(15,5))

a=s.barplot(x='Product line',y='Total',hue='Gender',data=df,order=l1,hue_order=['Male','Female'],palette='cividis',estimator=n.mean)
```
