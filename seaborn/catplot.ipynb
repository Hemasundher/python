# catplot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

basci catplot

```python
s.catplot(data=df,x='Payment',y='Total',kind='violin')
```

```python
s.catplot(data=df,x='Payment',y='Total',hue='Gender',kind='violin')
```

```python
s.catplot(data=df,x='Payment',y='Total',col='Gender',kind='violin')
```

```python
s.catplot(data=df,x='Payment',y='Total',col='Gender',kind='violin')
```

```python
s.catplot(data=df,x='Payment',y='Total',col='Gender',kind='violin',row='Customer type')
```

```python
s.catplot(data=df,x='Payment',y='Total',col='Gender',kind='bar',row='Customer type')
```
