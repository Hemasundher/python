# swarm plot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

basic swarm plot

```python
s.swarmplot(data=df,y='Total')
```

categoric value

```python
s.swarmplot(data=df,x='Payment',y='Total',palette='CMRmap')
```

```python
s.swarmplot(data=df,x='Payment',y='Total',palette='CMRmap',hue='Gender')
```

hues separetly on axis

```python
s.swarmplot(data=df,y='Total',x='Payment',hue='Gender',split=True)
```

marker size,color

```python
plt.figure(figsize=(15,5))
s.swarmplot(data=df,y='Total',x='Payment',marker='*',size=10,color='green',edgecolor="red")
```

swarm plot -- boxplot

```python
s.swarmplot(data=df,y='Total',x='Payment')
s.boxplot(data=df,x='Payment',y='Total',color='black')
```

swarmplot - violin plot

```python
s.swarmplot(data=df,y='Total',x='Payment',color='white')
s.violinplot(data=df,x='Payment',y='Total',inner=None)
```
