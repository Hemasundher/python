# displot - distribution plot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

```python
df=df[['Gender','Payment','Unit price','Quantity','Total','gross income','Branch']]
df.head()
```

displot - histogram

```python
s.displot(data=df,x='Total')
```

```python

s.histplot(data=df,x='Total')
```

```python
s.displot(data=df,x='Total',rug=True,rug_kws={'height':0.05},kde=True,kde_kws={'bw_adjust':1},hue='Payment',multiple='stack')
```

```python
s.displot(data=df,x='Total',rug=True,rug_kws={'height':0.05},kde=True,kde_kws={'bw_adjust':1},hue='Payment',multiple='stack',col='Gender')
```

```python
s.displot(data=df,x='Total',rug=True,rug_kws={'height':0.05},kde=True,kde_kws={'bw_adjust':1},hue='Payment',multiple='stack',row='Gender',col='Branch')
```

ecdf

```python
s.displot(data=df,x='Total',hue="Payment",kind='ecdf',row='Gender',col='Branch')
```
