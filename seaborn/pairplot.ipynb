# pairplot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
iris=s.load_dataset('iris')
iris.head()
```

basic pair plot

```python
s.pairplot(data=iris)
```

diagnol plot kind to kde

```python
s.pairplot(data=iris,diag_kind='kde')
```

```python
s.pairplot(data=iris,diag_kind='hist')
```

```python
s.pairplot(data=iris,diag_kind=None)
```

change non diagnol plot kind to scatter ,kde,hit or reg

```python
s.pairplot(data=iris,kind='reg')
```

adding hue to paiplot

```python
s.pairplot(data=iris,diag_kind='kde',hue='species')
```

creating a pairplot for specific list 

```python
s.pairplot(data=iris,x_vars=['sepal_width','petal_width','petal_length'],y_vars=['sepal_length'])
```

```python
s.pairplot(data=iris,x_vars=['sepal_width','petal_width','petal_length'],y_vars=['sepal_length','sepal_width'])
```

```python
s.pairplot(data=iris,x_vars=['sepal_width','petal_width','petal_length'],y_vars=['sepal_length','sepal_width'],diag_kind=None)
```

showing only lower traingle using corner argument

```python
s.pairplot(data=iris,corner=True)
```

making changes specific to diagnol and non diagnol plots searately

```python
s.pairplot(data=iris,diag_kws=dict(color='green',kde=True),
            plot_kws=dict(color='red',marker=10,s=100)
            )
```

kde plot on pairplot

```python
a=s.pairplot(data=iris,plot_kws=dict(color='red'))
a.map_upper(s.kdeplot)
a.map_lower(s.kdeplot)
```
