# stripplot

```python
import pandas as p
import numpy as n
import seaborn as s
import matplotlib.pyplot as plt
```

```python
df=p.read_csv('supermarket_sales.csv')
df.head()
```

basic strip plot

```python
s.stripplot(data=df,y='Total')
```

jitter 

```python
s.stripplot(data=df,y='Total',jitter=0.2)
```

linewidth

```python
s.stripplot(data=df,y='Total',jitter=0.3,linewidth=0.7)
```

one categorical value

```python
s.stripplot(data=df,y='Total',x='Payment',jitter=0.2,palette='CMRmap',linewidth=0.5)
```

one numeric and two categoric

```python
s.stripplot(data=df,y='Total',x='Payment',hue='Gender',linewidth=0.5,jitter=0.2)
```

separaing gender using dodge

```python
s.stripplot(data=df,x='Payment',y='Total',hue='Gender',linewidth=0.4,dodge=True,jitter=0.2)
```

stripplot and violinplot

```python
s.stripplot(data=df,x='Payment',y='Total',palette='CMRmap',jitter=0.1)
s.violinplot(data=df,x='Payment',y='Total',color='grey')
```

stripplot - boxplot

```python
s.stripplot(data=df,x='Payment',y='Total',palette='CMRmap')
s.boxplot(data=df,x='Payment',y='Total',color='grey')
```
